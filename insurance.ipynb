# Clean and fix data:

### we should remove the duplicated row

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

we remove it successfully

### we should remove outliers

#### Charges outliers

```python
charges_outliers = df[ df['charges'] >= 50000 ].count() 
charges_outliers
```

7 for 1338 is not a large number, it will not affect when droped

```python
charges_outliers = df[df['charges'] >= 50000 ].index
df.drop(charges_outliers , inplace=True)
```

```python
sns.boxplot(x=df['charges']);
```

The ourliers that exist btween 32500 and 50000 it is natural to exist, which does not require their removal.

#### BMI outliers

```python
bmi_outliers = df[df['bmi'] >= 46 ].index
df.drop(bmi_outliers , inplace=True)
```

```python
sns.boxplot(x=df['bmi']);
```

we remove it successfully
